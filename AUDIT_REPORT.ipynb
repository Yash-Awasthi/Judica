## Key Findings

### What Was Fixed (Critical Impact)

1. **Sandbox env leak** (SEC-2) — Python sandbox no longer receives full `process.env`. Only safe, explicitly listed variables are passed.

2. **eval()/new Function() injection** (SEC-1, WF-3) — All expression evaluation now uses a safe recursive descent math parser. No dynamic code compilation on user/agent input.

3. **SSRF protection** (WF-1, WF-6, PRV-5) — All outbound HTTP calls in workflow nodes, tools, and adapters now validate URLs against `lib/ssrf.ts` blocklist (private IPs, cloud metadata, localhost).

4. **WebSocket impersonation** (BE-3) — WebSocket connections now validate userId from the JWT token on the server side, preventing users from impersonating others.

5. **Hardcoded credentials** (INF-1, BE-4) — Docker Compose uses env var references instead of plaintext passwords. Admin passwords use per-user random salts.

6. **API key exposure** (PRV-1, PRV-2, BE-16) — Gemini API keys moved from URL query strings to request headers. SerpAPI keys moved to headers. Custom adapter auth secured.

7. **XSS vulnerability** (FE-1) — `dangerouslySetInnerHTML` in EnhancedSearch replaced with sanitized HTML rendering.

8. **Express migration completed** (DEP-1) — Last 2 Express route files (`ask.ts`, `uploads.ts`) migrated to native Fastify plugins. The migration claim is now genuinely true.

9. **Null guard bug fix** (Branch Audit) — `archetypes.ts` now guards against undefined `name`/`systemPrompt` before checking `.length`, preventing runtime TypeError crashes.

10. **Cost warning bug fix** (Branch Audit) — `realtimeCost.ts` now includes the current session's accumulated cost when checking daily/monthly warning thresholds. Previously, warnings wouldn't fire until the session was flushed to the ledger.

### What Still Needs Work

1. **Test coverage at 28.74%** — Improved from ~6% with the addition of 43 test files (335 tests, 100% pass rate), but still far from the 80% target. Entire modules at 0%: routes, adapters, workflow, tools, strategies, processors, sandbox.

2. **Redis-backed rate limiting** — In-memory rate limiters work for single-instance but will be bypassed in multi-instance deployments.

3. **Circuit breaker wiring** — The breaker code exists in `lib/breaker.ts` but isn't wired into all adapter call paths.

4. **HNSW vector indexes** — pgvector search uses sequential scan. For production-scale similarity queries, HNSW indexes are needed.

5. **Vitest version compatibility** — Project uses Vitest 4.x which deprecated `poolOptions` nesting. Config has been corrected, but `@vitest/coverage-v8` needs to be added to devDependencies.
